### Í∑∏ÎûòÌîÑ Ï†úÎ™© Ï∂îÍ∞ÄÌïòÍ∏∞

````python
plt.title('3Ïõî Í∏∞Ïò®') 
plt.xlabel('ÎÇ†Ïßú')
plt.ylabel('Í∏∞Ïò®')

plt.plot(temp)
plt.show()
````

### ÏÑ† Î™®Ïñë Ï†ïÌïòÍ∏∞
+ **ÏÉâÏÉÅ**
 - b(ÌååÎûÄÏÉâ), g(Ï¥àÎ°ùÏÉâ), r(Îπ®Í∞ÑÏÉâ), c(Ï≤≠Î°ùÏÉâ), y(ÎÖ∏ÎûÄÏÉâ), k(Í≤ÄÏùÄÏÉâ), w(Ìù∞ÏÉâ)
+ **ÎßàÏª§**
 - o(Ïõê), v(Ïó≠ÏÇºÍ∞ÅÌòï), ^(ÏÇºÍ∞ÅÌòï), s(ÎÑ§Î™®), +(ÌîåÎü¨Ïä§), .(Ï†ê)

````python
plt.plot(temp, 'o--')
plt.show()
````

### Í∑∏ÎûòÌîÑ ÌÅ¨Í∏∞ Ï°∞Ï†à
+ plt.figure() 
 - figsizeÎ•º Ïù¥Ïö©ÌïòÏó¨ Í∞ÄÎ°ú, ÏÑ∏Î°ú Í∏∏Ïù¥ Ï°∞Ï†à Í∞ÄÎä• (inch Îã®ÏúÑ)

````python
plt.figure(figsize=(16,9))

plt.title('3Ïõî Í∏∞Ïò®') 
plt.xlabel('ÎÇ†Ïßú')
plt.ylabel('Í∏∞Ïò®')

plt.plot(temp, 'go--')
plt.show()
````

### Í≤©Ïûê Î∞è ÏàòÌèâ/ÏàòÏßÅÏÑ† Ï∂îÍ∞ÄÌïòÍ∏∞
#### Í∑∏Î¶¨Îìú (grid)
+ grid() Ìï®ÏàòÏùò Ï≤´Î≤àÏß∏ ÌååÎùºÎØ∏ÌÑ∞Î•º TrueÎ°ú ÏÑ§Ï†ïÌïòÎ©¥ Í∑∏ÎûòÌîÑÏóê Í∑∏Î¶¨Îìú (grid)Í∞Ä ÌëúÏãú 
+ alpha:  Í∑∏Î¶¨ÎìúÏùò Ìà¨Î™ÖÎèÑÎ•º ÏÑ§Ï†ï
 - 0ÏúºÎ°ú ÏÑ§Ï†ïÌïòÎ©¥ Ìà¨Î™ÖÌïòÍ≤å, 1ÏùÄ Î∂àÌà¨Î™ÖÌïòÍ≤å ÌëúÏãú
+ linestyle : ÏÑ†Ïùò Ïä§ÌÉÄÏùºÏùÑ ÎåÄÏâ¨ (Dashed)Î°ú ÏÑ§Ï†ï 


#### ÏàòÌèâ / ÏàòÏßÅÏÑ† Í∑∏Î¶¨Í∏∞
+ plt.axvline(x, color)
 - Ï∂ïÏùÑ Í∞ÄÎ°úÏßÄÎ•¥Îäî ÏÑ∏Î°ú ÏÑ† ÏÉùÏÑ±
+ plt.axhline(y, color)
 - Ï∂ïÏùÑ Í∞ÄÎ°úÏßÄÎ•¥Îäî Í∞ÄÎ°ú ÏÑ† ÏÉùÏÑ±

````python
plt.figure(figsize=(10,6)) 

plt.title('3Ïõî Í∏∞Ïò®') 
plt.xlabel('ÎÇ†Ïßú')
plt.ylabel('Í∏∞Ïò®')

plt.plot(temp, 'go--')
plt.grid(True, color='gray', alpha=0.5, linestyle='--')
plt.axvline(x=10, color='r', linewidth=1)
plt.axhline(y=8, color='r', linewidth=1)
plt.show()
````

### Í∏ÄÏûê Ïì∞Í∏∞
+ plt.text(x, y, s, fontsize)
 - ÏõêÌïòÎäî ÏúÑÏπòÏóê ÌÖçÏä§Ìä∏ ÏÉùÏÑ±

````python
plt.figure(figsize=(10,6)) 

plt.title('3Ïõî Í∏∞Ïò®') 
plt.xlabel('ÎÇ†Ïßú')
plt.ylabel('Í∏∞Ïò®')

plt.plot(temp, 'go--')
plt.grid(True, color='gray', alpha=0.5, linestyle='--')
plt.axhline(y=10, color='r', linewidth=1)

for i in range(len(temp)) :
  plt.text(i-0.2, temp[i]+0.2, temp[i])
plt.show()
````

### xÏ∂ï ÎààÍ∏à ÌëúÏãú Î∞è ÌöåÏ†Ñ

### subplot
+ ÌïòÎÇòÏùò figureÏóê Ïó¨Îü¨Í∞ú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞ 

````python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(temp, 'go--')
plt.title('3Ïõî Í∏∞Ïò®1') 

plt.subplot(1,2,2)
plt.plot(temp, 'bx')
plt.title('3Ïõî Í∏∞Ïò®2')

plt.show()
````

````python
fig = plt.figure(figsize=(10,4)) 
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2) 

ax1.plot(temp, 'go--')
ax1.set_title('3Ïõî Í∏∞Ïò®1') 

ax2.plot(temp, 'bx')
ax2.set_title('3Ïõî Í∏∞Ïò®2') 
plt.show()
````

### Îã§ÏñëÌïú Í∑∏ÎûòÌîÑ Í∑∏Î¶¨Í∏∞
+ bar : ÎßâÎåÄÍ∑∏ÎûòÌîÑ 
+ scatter : ÏÇ∞Ï†êÎèÑ Í∑∏ÎûòÌîÑ
+ pie : ÌååÏù¥ Í∑∏ÎûòÌîÑ
+ hist : ÌûàÏä§ÌÜ†Í∑∏Îû® Í∑∏ÎûòÌîÑ
+ boxplot : Î∞ïÏä§ÌîåÎ°Ø Í∑∏ÎûòÌîÑ
+ imshow() :  Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞Ï≤òÎüº ÌñâÍ≥º Ïó¥ÏùÑ Í∞ÄÏßÑ ÌñâÎ†¨ ÌòïÌÉúÏùò 2Ï∞®Ïõê Îç∞Ïù¥ÌÑ∞ Í∑∏ÎûòÌîÑ

````python
x = list(range(1,32))

plt.figure(figsize=(10,6))

plt.subplot(2,3,1)
plt.bar(x, temp)
plt.title('ÎßâÎåÄÍ∑∏ÎûòÌîÑ')
 
plt.subplot(2,3,2)
plt.barh(x,temp)
plt.title('ÎßâÎåÄÍ∑∏ÎûòÌîÑ')

plt.subplot(2,3,3)
plt.scatter(x, temp)
plt.title('ÏÇ∞Ìè¨ÎèÑÍ∑∏ÎûòÌîÑ')

plt.subplot(2,3,4)
plt.pie(temp)
plt.title('ÌååÏù¥ Í∑∏ÎûòÌîÑ')
 
plt.subplot(2,3,5)
plt.hist(temp)
plt.title('ÌûàÏä§ÌÜ†Í∑∏Îû®')

plt.subplot(2,3,6)
plt.boxplot(temp)
plt.title('Î∞ïÏä§ÌîåÎ°Ø')

plt.show()
````
